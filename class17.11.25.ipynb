```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.relplot(data=tips_df, x='total_bill', y='tip')
```

```python
sns.relplot(data=tips_df, x='total_bill', y='tip', hue='smoker')
plt.show()
```

```python
sns.relplot(data=tips_df, x='total_bill', y='tip', hue='smoker', style='smoker')
plt.show()
```

```python
sns.relplot(data=tips_df, x='total_bill', y='tip', hue='smoker', style='smoker', size='size')
plt.show()
```

```python
sns.relplot(data=tips_df, x='total_bill', y='tip', hue='smoker', style='smoker', size='size', col='smoker')
plt.show()
```
