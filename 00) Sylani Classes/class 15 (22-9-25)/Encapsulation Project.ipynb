```python
class Student():
    # Since institute and campus attrributes are common to all students, So i have declared them as class attributes
    # only administration have rights to modify these attributes to they are private class attributes
    __institute = "Saylani Mass IT (SMIT)"
    __campus = "Hajyani Sakeena Campus"
    
    # constructor
    def __init__(self, name, roll_no, course, marks, phone=None, email=None):
        # instance attributes (Private)
        # because name, roll_no, course, marks and contact details are sensitive, no one can change these attributes instead of Administration
        self.__name = name
        self.__roll_no = roll_no
        self.__course = course
        self.__marks = marks
        self.__contact_detail = {"Phone": phone, "Email": email}

    # Instance Methods-------------------------------------------------------
    # Because instance attributes are private only instace methods can 
    # access/modify these attributes, So i have used instance methods to 
    # access/modify these attributes
    # -------------------------Public methods -------------------------------
    # Getter methods

    # get name
    def get_name(self):
        return self.__name

    # get roll no
    def get_roll_no(self):
        return self.__roll_no

    # get course
    def get_course(self):
        return self.__course

    # get marks
    def get_marks(self):
        return self.__marks

    # get contact
    def get_contact(self):
        return self.__contact_detail

    # get institute
    def get_institute(self):
        return self.__institute

    # get campus
    def get_campus(self):
        return self.__campus

    # Setter Method
    # modify campus
    def set_campus(self, new_campus):
        self.__campus = new_campus 

    # modify marks
    def set_grade(self, update_marks):
        if Student.__validate_marks(update_marks):
            self.__marks = update_marks
        else:
            print("Invalid Marks!!!")

    # modify student contact
    def set_contact(self, updated_contact):
        if Student.__validate_contact(updated_contact):
            for contact_type, contact_info in updated_contact.items():
                if contact_type in self.__contact_detail:
                    self.__contact_detail[contact_type] = contact_info

    # ------------------------- Private Methods-----------------------------
    # Validate marks
    #def __validate_marks(self, validate_marks):
    # THis mehtod is not accessing instance attribute, but to check the marks. So I can declare this method as static method
    @staticmethod
    def __validate_marks(validate_marks):
        return isinstance(validate_marks, (int, float)) and 0 <= validate_marks <=100

    #def __validate_contact(self, validate_contact):
    @staticmethod
    def __validate_contact(validate_contact):
        phone = validate_contact.get("Phone")
        email = validate_contact.get("Email")
    
        # validate phone
        if phone and not str(phone).isdigit():
            # Invalid phone number
            return False

        # validate email
        if email and '@' not in email:
            # Invalid email
            return False

        return True
```

Now creating Students Records Management System Functionality Class  (Manager Class)

```python
class StudentsRecordsManagementSystem():
    # constructor
    def __init__(self):
        # for storing students objects (Private)
        self.__students = {}      # key --> roll_no and values --> studeent object

    # Instance Methods
    def dict_to_str(self):
        return {key: value for key, value in self.__students.items()}
                  
    def __str__(self):
        return str(self.dict_to_str())

    # ---------------- Display Methods ----------------------
    @staticmethod
    def display_contact(contact):
        print("Contact Details -------------------------")
        for contact_type, contact_info in contact:
            print(f"{contact_type}: {contact_info}")            
        print("-----------------------------------------")


    # display student method     -------------------------   Step 5
    @staticmethod
    def display_student(roll_no, student):
        if isinstance(student, Student):
            print("========== Institute Details ============")
            print(f"Institute: {student.get_institute()}")
            print(f"Campus: {student.get_campus()}")
            print("=========================================")
            print(f"Name: {student.get_name()}")
            print(f"Roll No: {student.get_roll_no()}")
            print(f"Marks: {student.get_marks()}")
            print(f"Course: {student.get_course()}")
            print("=========================================")
            #self.display_contact(student.get_contact().items())
            print("Contact Details -------------------------")
            for contact_type, contact_info in student.get_contact().items():
                print(f"{contact_type}: {contact_info}")            
            print("-----------------------------------------")

        else:
            print("=========================================")
            print(f"{roll_no} Record Deleted...")
            print("=========================================")

    # ---------------- Input Methods -----------------------
    # take inputs    -------------------------   Step 2
    def input_student_details(self):
        name = input("Enter name: ")
        roll_no = input("Enter roll no: ")
        course = input("Enter Course: ")
        marks = float(input("Enter Marks: "))
        phone_no = input("Enter Phone no: ")
        email = input("Enter email address: ")

        if roll_no not in self.__students:
            return StudentsRecordsManagementSystem.create_student_object(name, roll_no, course, marks, phone_no, email)
        else:
            return None

    # input contact details
    @staticmethod
    def input_contact():
        print("-------------------------------")
        print("Enter {p} to update phone no: ")
        print("Enter {e} to update email address: ")
        print("-------------------------------")

        update_choice = input("Enter your choice: ").strip().lower()
        if update_choice == "p":
            new_phone = input("Enter Phone: ")
            return {"Phone": new_phone}
        elif update_choice == "e":
            new_email = input("Enter Email address: ")
            return {"Email": new_email}
        else:
            print("Invalid Choice")
            return None

    @staticmethod
    def input_roll_no():
        rollno = input("Enter Roll no:")
        return rollno

    @staticmethod
    def input_update_marks():
        marks = float(input("Enter marks to update: "))
        return marks 

    @staticmethod
    def input_new_campus():
        campus = input("Enter new campus: ")
        return campus
    
    # create student object    -------------------------   Step 3
    @staticmethod
    def create_student_object(name, roll_no, course, marks, phone_no, email):
        if phone_no == '' and email == '':
            student_obj = Student(name, roll_no, course, marks)
        elif phone_no == '':
            student_obj = Student(name, roll_no, course, marks, email=email)
        elif email == '':
            student_obj = Student(name, roll_no, course, marks, phone=phone_no)
        else:
            student_obj = Student(name, roll_no, course, marks, phone=phone_no, email=email)

        return student_obj
            
    # Add student    -------------------------   Step 1
    def add_student(self):
        student = self.input_student_details()
        if isinstance(student, Student):
            self.__students[student.get_roll_no()] = student 

            print(f"Student {student.get_name()} added successfully...")
        else:
            print("Student already exist")        
        
    # Searching Student    -------------------------   Step 6
    def search_student(self):
        roll_number = StudentsRecordsManagementSystem.input_roll_no()
        student = self.__students.get(roll_number, None)
        if student is not None:
            self.display_student(roll_number, student)
            return student     # must have student object
        else:
            print(f"{roll_number} does not exist in the records")
            return student     # No student object (i.e. None)

    # Modify campus    -------------------------   Step 11
    def change_campus(self):
        student = self.search_student()
        if student:
            new_campus = StudentsRecordsManagementSystem.input_new_campus()
            student.set_campus(new_campus)
            print(f"New Campus {new_campus} for student {student.get_name()} updated_successfully")
        else:
            print(f"Student not exist... Unable to update campus!!!")
        
    # Update Student marks    -------------------------   Step 7
    def update_student(self):
        student = self.search_student()
        if student:
            update_marks = StudentsRecordsManagementSystem.input_update_marks()
            student.set_grade(update_marks)
            print(f"Marks {update_marks} for student {student.get_name()} updated successfully...")
        else:
            print(f"Student not exist... Unable to update marks!!!")

    # modify contact    -------------------------   Step 8
    def update_contact(self):
        student = self.search_student()
        if student:
            # display contact
            StudentsRecordsManagementSystem.display_contact(student.get_contact().items())
            updated_contact = StudentsRecordsManagementSystem.input_contact()
            if updated_contact:
                student.set_contact(updated_contact)

        else:
            print("Unable to update contact!!!")

    # delete student    -------------------------   Step 9
    def delete_student(self):
        student = self.search_student()
        if student:
            roll_number = student.get_roll_no()
            del student
            self.__students[roll_number] = "Deleted_record..."
            print(f"{roll_number} deleter successfully")

    # display students    -------------------------   Step 4
    def display_all(self):
        if self.__students:
            for roll_no, student_obj in self.__students.items():
                self.display_student(roll_no, student_obj)
                #print("=====================================")
                #print(f"Name: {student_obj.get_name()}")
                #print(f"Roll: {student_obj.get_roll_no()}")
                #print(f"Marks: {student_obj.get_marks()}")
                #print(f"Course: {student_obj.get_course()}")
                #print("=====================================")
        else:
            print("No Records to display")

    # Sort students    -------------------------   Step 10
    def sort_records(self):
        self.__students = dict(
                            sorted(self.__students.items(),
                            key=lambda item:(1 if isinstance(item[1], Student) else 0,
                                              item[1].get_marks() if isinstance(item[1], Student) else 0),
                            reverse=True
                            )
                        )
        print("Records sorted successfully...")
```
